```
#
#    
#

def cosine_similarity(v, w):
    return np.dot(v, w) / math.sqrt(np.dot(v, v) * np.dot(w, w))
```

```
unique_books = sorted(list({ book
                                 for book in df['Book-Title']
                                 for books in df['Book-Title'] }))
```

```
#     
def make_user_books(user_books):
    """       ,
     i-   1,  unique_interests[i]   ,
     0   """
    return [1 if book in user_books else 0
            for book in unique_books]
```

```
books_similarities = [[cosine_similarity(book_vector_i, book_vector_j)
                      for book_vector_j in user_books_matrix]
                     for book_vector_i in user_books_matrix]
```

```
def most_similar_books(book_id):
    pairs = [(other_book_id, similarity)                      # find other
             for other_book_id, similarity in                 # users with
                enumerate(books_similarities[book_id])         # nonzero
             if book_id != other_book_id and similarity > 0]  # similarity

    return sorted(pairs,                                      # sort them
                  key=lambda pair: pair[1],                   # most similar
                  reverse=True)                               # first
```

```
def book_based_suggestions(book_id, include_current_books=False):
    # sum up the similarities
    suggestions = defaultdict(float)
    for other_book_id, similarity in most_similar_books(book_id):
        for interest in books_interests[other_book_id]:
            suggestions[interest] += similarity

    # convert them to a sorted list
    suggestions = sorted(suggestions.items(),
                         key=lambda pair: pair[1],
                         reverse=True)

    # and (maybe) exclude already-interests
    if include_current_books:
        return suggestions
    else:
        return [(suggestion, weight)
                for suggestion, weight in suggestions
                if suggestion not in books_interests[book_id]]
```
